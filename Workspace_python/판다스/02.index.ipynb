```python
data={
  '이름':['김자바', '이자바', '박자바', '최자바', '정자바'],
  '학급':['자바반', '자바반', '자바반', '캐드반', '캐드반'],
  '국어점수':[90,80,70,100,90],
  '영어점수':[70,80,70,80,90],
  '평균':[80.0,80.0,75.0,85.0,90.0],
  '취미':['독서', '', '', '요가', '축구']
}
```

```python
import pandas
```

```python
# 데이터 프레임 생성
df=pandas.DataFrame(data)
```

```python
# 데이터 프레임 생성시 index지정
df=pandas.DataFrame(data, index=['1번', '2번', '3번', '4번', '5번'])

# 인덱스의 이름을 설정
df.index.name='학번'

df
```

```python
# 인덱스를 초기화
# 파이썬은 트루 펄스 무조건 대문자 앞에 
# drop=true :index 가 지워지는 효과
# drop=true 작성하지 않으면 인덱스가 컬럼으로 변경
# 인덱스 초기화를 저장하기위해서는 index=true 작성

df.reset_index(drop=True, inplace=True)
```

```python
# 새로운 내용을 확인하기 위해 데이터프레임을 새로 생성
df=pandas.DataFrame(data)
df
```

```python
# 컬럼을 index로 지정
# 이름을 index로 지정
# inplace =true  를 작성해야 저장이 됨
df.set_index('이름', inplace=True)
```

```python
df
```

```python
# 위에서 변경한 인덱스를 초기화
# inplace=True   완전 확정하는것
df.reset_index(drop=True)
```

```python
# 위에서 변경한 인덱스 다시 초기화
df.reset_index(inplace=True)
```
